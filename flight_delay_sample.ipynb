```python
# load data & check fore issue
import pandas as pd
```

```python

df = pd.read_csv("/content/Combined_Flights_2018.csv")
df.info()
df.head()
df.describe()
```

```python
#Handling missing Values
df.fillna({"Airline":"Unknown", "Origin": "Unknown", "Dest": "Unknown"}, inplace =True)

df["DepDelay"].fillna(df["DepDelay"].mean(), inplace = True)

df.dropna(subset=["CRSDepTime", "CRSArrTime"], inplace = True)

df.info()
```

```python
#3. CONVERTING TIME FIELD TO USABLE FEATURES
df["CRSDepTime"] = pd.to_datetime(df["CRSDepTime"], format = "%H%M", errors="coerce")
df["HourOfDay"] =df["CRSDepTime"].dt.hour
df["IsWeekend"] = df["DayOfWeek"].apply(lambda x: 1 if x in [6,7] else 0)
```

```python
# 4. Encode Categorical Features

from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
df["Airline_Index"] = encoder.fit_transform(df["Airline"])
df["Origin_Index"] = encoder.fit_transform(df["Origin"])
df["Dest_index"] = encoder.fit_transform(df["Dest"])
df.head()
```

```python
# FEATURE SCALING For Numerical columns

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df[["DepDelay","ArrDelay","TaxiOut","TaxiIn", "Distance"]] = scaler.fit_transform(df[["DepDelay","ArrDelay","TaxiOut","TaxiIn", "Distance"]])
df[["DepDelay","ArrDelay","TaxiOut","TaxiIn", "Distance"]].head()
```

```python
#  FEATURE SELECTION (Droping Irrelevant Columns)


columns_to_drop = ["FlightNumber", "Tail_Number", "DOT_ID_Marketing_Airline",
                   "Operated_or_Branded_Code_Share_Partners", "IATA_Code_Operating_Airline",
                   "DivAirportLandings", "DepTimeBlk", "ArrTimeBlk", "OriginState", "DestState"]

df = df.drop(columns=columns_to_drop, axis=1)

```

```python
# Final DATACHECK
df.info()
df.head()
```

```python
# SAVING PREPROCESSED DATA
df.to_csv("cleaned_flight_delay_data.csv", index=False)
```
